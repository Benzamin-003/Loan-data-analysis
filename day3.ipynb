# Section 3: Encoding Categorical Variables

# 15. Which columns need encoding before applying machine learning models? Why? 

 person_gender, person_education, 
 person_home_ownership, loan_intent, previous_loan_defaults_on_file needs encoding before applying machine learning models.

```python
df['person_gender'] = df['person_gender'].map({'male': 1, 'female': 0})
df = pd.get_dummies(df, columns=['person_home_ownership'], drop_first=True, dtype='int')
df = pd.get_dummies(df, columns=['loan_intent'], drop_first=True, dtype='int')
```

```python
edu_lvl = {
    'High School': 0,
    'Associate': 1,
    'Bachelor': 2,
    'Master': 3,
    'Doctorate': 4}
df['person_education'] = df['person_education'].replace(edu_lvl)
df
```
